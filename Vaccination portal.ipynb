```python
def login_portal():
    attempts = 3
    while attempts > 0:
        login_id = input('Enter your login id: ')
        password = input('Enter your password: ')
        if login_id == 'Ahil' and password == '123':
            print(f'Welcome {login_id} login successfuly')
            break
        else:
            attempts -= 1
            if attempts > 0:
                print(f"Wrong login ID or password. You have {attempts} attempts left.")
            else:
                print("Login failed. Maximum attempts exceeded.")
login_portal()
```
